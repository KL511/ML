```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
wine = fetch_ucirepo(id=109)

# data (as pandas dataframes)
X = wine.data.features
y = wine.data.targets
Z = pd.concat([X.head(),X.tail()])
Z

#вино разделено на 3 сорта по 13 качествам
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn import metrics
X_train_corr = X_train.loc[:,'Total_phenols':'Flavanoids']
X_test_corr = X_test.loc[:,'Total_phenols':'Flavanoids']
model = DecisionTreeClassifier()
model.fit(X_train_corr, y_train)
y_predict = model.predict(X_test_corr)
res_f1score = f1_score(y_test, y_predict, average = 'macro')
res_accuracy = accuracy_score(y_test, y_predict)
res_precision = precision_score(y_test, y_predict, average = 'macro')
res_recall = recall_score(y_test, y_predict, average = 'macro')
res_cmatrix = metrics.confusion_matrix(y_test, y_predict)
print('f1 score:',res_f1score,',','accuracy:',res_accuracy,',','precision:',res_precision,',','recall:',res_recall)
sns.heatmap(res_cmatrix, annot=True, linewidth=5, cbar=None)
plt.ylabel('Real label')
plt.xlabel('Predicted label')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, confusion_matrix
X_train_anticorr = pd.concat([X_train.loc[:,'Alcalinity_of_ash'],X_train.loc[:,'Color_intensity']], axis = 1)
X_test_anticorr = pd.concat([X_test.loc[:,'Alcalinity_of_ash'],X_test.loc[:,'Color_intensity']], axis = 1)
model = DecisionTreeClassifier()
model.fit(X_train_anticorr, y_train)
y_predict = model.predict(X_test_anticorr)
res_f1score = f1_score(y_test, y_predict, average = 'macro')
res_accuracy = accuracy_score(y_test, y_predict)
res_precision = precision_score(y_test, y_predict, average = 'macro')
res_recall = recall_score(y_test, y_predict, average = 'macro')
res_cmatrix = confusion_matrix(y_test, y_predict)
print('f1 score:',res_f1score,',','accuracy:',res_accuracy,',','precision:',res_precision,',','recall:',res_recall)
sns.heatmap(res_cmatrix, annot=True, linewidth=5, cbar=None)
plt.ylabel('Real label')
plt.xlabel('Predicted label')
```

```python
from sklearn.ensemble import RandomForestClassifier
model_RF = RandomForestClassifier(max_depth=2, random_state=0)
model_RF.fit(X_train, y_train)
y_predict = model_RF.predict(X_test)
res_f1score = f1_score(y_test, y_predict, average = 'macro')
res_accuracy = accuracy_score(y_test, y_predict)
res_precision = precision_score(y_test, y_predict, average = 'macro')
res_recall = recall_score(y_test, y_predict, average = 'macro')
res_cmatrix = confusion_matrix(y_test, y_predict)
print('f1 score:',res_f1score,',','accuracy:',res_accuracy,',','precision:',res_precision,',','recall:',res_recall)
sns.heatmap(res_cmatrix, annot=True, linewidth=5, cbar=None)
plt.ylabel('Real label')
plt.xlabel('Predicted label')
```

```python
importances = model_RF.feature_importances_
# Сортировка индексов признаков по убыванию важности
indices = np.argsort(importances)[::-1]
plt.figure(figsize=(10, 6))
plt.title("Feature importances")
plt.bar(range(X_train.shape[1]), importances[indices], color="b", align="center")
plt.xticks(range(X_train.shape[1]), [X_train.columns.tolist()[i] for i in indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.show()
```

```python
y_prob = model_RF.predict_proba(X_test)
# Для каждого класса строим ROC-кривую
plt.figure(figsize=(10, 10))
for i in range(3):
    fpr, tpr, thresh = metrics.roc_curve(y_test, y_prob[:, i], pos_label=model_RF.classes_[i])
    auc_roc = metrics.auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'ROC curve (class {i}, area = {auc_roc:.2f})')

plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('FP rate')
plt.ylabel('TP rate')
plt.title('ROC')
plt.legend(loc='lower right')
```
